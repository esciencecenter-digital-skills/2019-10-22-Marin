```python
gdps = [ col for col in data.columns if col.startswith('gdp') ]
life = [ col for col in data.columns if col.startswith('life') ]
pop = [ col for col in data.columns if col.startswith('pop') ]
gdps
```

```python
plt.scatter(data.loc['Australia', gdps], data.loc['Australia', life])
```

```python
plt.scatter(data.loc[:, gdps], data.loc[:, life])
plt.xscale('log')
```

```python
plt.scatter(data.loc[:, gdps], data.loc[:, life], s=data.loc[:, pop]**0.1, )
plt.xscale('log')
```

```python
data['color'] = data['continent'].map({
    'Africa': 'r',
    'Americas': 'g',
    'Europe': 'b',
    'Asia': 'm',
    'Oceania': 'y'
})
```

```python
plt.scatter(data.loc[:, 'gdpPercap_1952'], data.loc[:, 'lifeExp_1952'], 
            s=data.loc[:, 'pop_1952']**0.1, c=data.loc[:, 'color'])
plt.xscale('log')
```
